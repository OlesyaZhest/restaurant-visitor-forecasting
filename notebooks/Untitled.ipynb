```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Импортируем данные и сделаем дату индексом.

```python
air_reserve = pd.read_csv(
    "C:/Users/Jessy/Desktop/study/ml/jupyter/homework/data/air_reserve.csv",
    parse_dates=["reserve_datetime", "visit_datetime"],
)
air_reserve.head()
```

```python
daily_visit = (
    air_reserve.groupby(air_reserve["visit_datetime"].dt.date)["reserve_visitors"]
    .sum()
    .reset_index()
)
daily_visit.columns = ["visit_datetime_day", "sum_visitors"]
```

```python
daily_visit.head()
```

```python
daily_visit.set_index("visit_datetime_day", inplace=True)
daily_visit.head()
```

```python
daily_visit.index = pd.to_datetime(daily_visit.index)
daily_visit.index[:5]
```

```python
ax = daily_visit.plot(
    xlabel="visit_datetime_day", ylabel="sum_visitors", figsize=(12, 6)
)
ax.set(xlabel="visit_day", ylabel="visitors")
```

```python
custom_dates = ["2016-01-01", "2016-04-10", "2016-07-19", "2016-10-27", "2017-02-04"]
custom_dates = pd.to_datetime(custom_dates)

# Построение графика
plt.figure(figsize=(12, 6))
ax = daily_visit.plot(
    xlabel="visit_datetime_day", ylabel="sum_visitors", figsize=(12, 6)
)
ax.set(xlabel="visit_day", ylabel="visitors")

# Установка меток на оси X
plt.xticks(custom_dates, custom_dates.strftime("%Y-%m-%d"), rotation=45)

plt.tight_layout()
plt.show()
```
